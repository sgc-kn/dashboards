```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV-Datei laden
df = pd.read_csv("maja_geodaten\\konstanz_flaechenanalyse4.csv")  # Pfad ggf. anpassen

# Gewnschte Reihenfolge der Flchentypen
categories_ordered = ["asphalt_%", "kies_%", "gebaeude_%", "baeume_%", "gruen_%", "wasser_%"]

# Umstrukturieren der Daten ins Long-Format
df_melted = df[["name"] + categories_ordered].melt(
    id_vars="name", 
    var_name="Flchentyp", 
    value_name="Flche in %"
)

# Schne Bezeichnungen fr die Legende
label_map = {
    "asphalt_%": "Asphalt",
    "kies_%": "Kies",
    "gebaeude_%": "Gebude",
    "baeume_%": "Bume",
    "gruen_%": "Grnflche",
    "wasser_%": "Wasser"
}
df_melted["Flchentyp"] = df_melted["Flchentyp"].map(label_map)

# Benutzerdefinierte Farbpalette
custom_palette = {
    "Asphalt": "#4D4D4D",      # Dunkelgrau
    "Kies": "#A9A9A9",         # Hellgrau
    "Gebude": "#B36A5E",#FFCB77",      # sunset
    "Bume": "#006400",        # Dunkelgrn
    "Grnflche": "#9BC53D",#87BBA2",   # Helles Grn
    "Wasser": "#1E90FF"        # Krftiges Blau
}

# Seaborn-Balkendiagramm zeichnen
plt.figure(figsize=(14, 6))
sns.barplot(
    data=df_melted,
    x="name",
    y="Flche in %",
    hue="Flchentyp",
    palette=custom_palette
)


# Vertikale gestrichelte Linien zwischen Stationen setzen
stationen = df["name"].unique()
for i in range(1, len(stationen)):
    plt.axvline(x=i - 0.5, color="gray", linestyle="--", linewidth=0.7)


plt.axhline(y= 50, color="gray", linestyle="--", linewidth=0.7)

# Beschriftung & Layout
plt.xticks(rotation=45, ha="right")
plt.title("Flchenanteile pro Station", fontsize="18", fontweight="bold")
plt.ylabel("Flche in %", fontsize="14")
plt.xlabel("Stationen", fontsize="14")

plt.legend(title="Flchentyp")
plt.tight_layout()

plt.show()
```
